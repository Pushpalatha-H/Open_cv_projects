# Image conversion project colored image into grayscale

```python
import cv2
from google.colab.patches import cv2_imshow
import numpy

path = input("Enter the path of the image==")
print("You entered this===", path)

# read the image
img1 = cv2.imread(path,0)               # convert the image into gray scale
img1 = cv2.resize(img1,(560, 700))       # width, height

print("gray image\n", img1)
cv2_imshow(img1)                                      # to show the img


k = cv2.waitKey(0)
if k==ord("s"):
  cv2.imwrite("C:\output.png", img1)
else:
  cv2.destroyAllWindows()
```
